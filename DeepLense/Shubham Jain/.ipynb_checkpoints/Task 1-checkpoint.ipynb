PyAutolens allows one to simulate **Strong Gravitational Lensing** from _galaxies_ with various mass and light profiles. It even includues mass profiles for dark matter halos in galaxies such as the _NFW profile_. It however, _lacks_ a convininent interface to simulate the strong gravitational lensing from other types of dark matter structures such as **Dark Matter Filaments** or strings of dark matter that usually connect galaxies and forms the large-scale structure of the universe. 

**Below is my approach to simulate the strong gravitational lensing caused by superfluid dark matter filaments with a constant mass density:**

# 1. Importing the required python modules

```python
%matplotlib inline

import autolens as al
import autolens.plot as aplt

import numpy as np
import random
import math
```

# 2. Custom dark matter filament implementation
Since PyAutolens lacks a convinient interface to create dark matter filament structures, I have defined a class that uses multiple point masses to simulate the same.

```python
class DarkMatterFilament():
    '''
        Class to create a thread of dark matter
        ------------------------------------------
        parameters:
            centre -> centre of dark matter filament in image plane
            length -> length of dark matter from centre to end, in arcsecs
            einstein_radius -> einstein_radius of each point mass in filament. Indirect measure of filament mass
            angle -> angle in degrees of filament with respect to the image plane's x-axis
            jitter -> set to True if filament should not be a straight line
    '''
    
    # setting default values of parameters
    centre = (0, 0)
    length = 7
    einstein_radius = 0.5
    angle = 0
    jitter = False
    
    # data structures to store all point masses that compose the thread
    mass_profiles_list = []
    mass_profiles_dict = {}
        
    def __init__(self, centre = (0, 0), length = 7, einstein_radius = 0.5, angle = 0, jitter = False):
        '''
            Function to initialize DarkMatterFilament class object
            -----------------------------------------------------------
        '''
        self.centre = centre
        self.length = length
        self.einstein_radius = einstein_radius
        self.angle = angle % 360
        self.jitter = jitter
        
        # add a point mass at chosen centre
        self.mass_profiles_list = [al.mp.PointMass(centre = self.centre, einstein_radius = self.einstein_radius)]
        # determine distance from centre for each point mass
        locations = np.arange(0.25, self.length, 0.25)

        for i in locations:
            if(self.jitter):
                # random jitter values dependent on filament orientation
                jy1 = random.uniform(-0.2, 0.2) * math.sin(math.radians(self.angle - 90))
                jx1 = random.uniform(-0.2, 0.2) * math.sin(math.radians(self.angle))
                jy2 = random.uniform(-0.2, 0.2) * math.sin(math.radians(self.angle - 90))
                jx2 = random.uniform(-0.2, 0.2) * math.sin(math.radians(self.angle))
            else:
                jy1, jx1, jy2, jx2 = (0, 0, 0, 0)
            
            # calculate co-ordinates of point masses, symmetric about the centre    
            y1 = self.centre[0] + (i * math.sin(math.radians(self.angle))) + jy1
            x1 = self.centre[1] + (i * math.cos(math.radians(self.angle))) + jx1
            y2 = self.centre[0] - (i * math.sin(math.radians(self.angle))) + jy2
            x2 = self.centre[1] - (i * math.cos(math.radians(self.angle))) + jx2
            
            # add calculated point masses
            self.mass_profiles_list.append(al.mp.PointMass(centre = (y1, x1), 
                                                           einstein_radius = self.einstein_radius))
            self.mass_profiles_list.append(al.mp.PointMass(centre = (y2, x2), 
                                                           einstein_radius = self.einstein_radius))
        
    def getMassProfilesList(self):
        '''
            Function to return a list of created point masses
            ---------------------------------------------------------
        '''
        return self.mass_profiles_list
    
    def getKeywordArgs(self):
        '''
            Function to return a dictionary of created point masses which can be 
            used as keyword arguments for masses when creating a 'galaxy'.
            -----------------------------------------------------------------------
        '''
        for i, mass_profile in enumerate(self.mass_profiles_list):
            self.mass_profiles_dict['mass' + str(i)] = mass_profile
        return self.mass_profiles_dict
```

# 3. Utility functions 

```python
def combine(darkMatterFilamentList):
    '''
        Function to combine a all point masses in list of DarkMatterFilament objects.
        --------------------------------------------------------------------------------
        input: darkMatterFilamentList -> list of darkMatterFilament objects
        output: list of all point masses in given darkMatterFilament objects
    '''
    result = []
    for darkMatterFilament in darkMatterFilamentList:
        result += darkMatterFilament.getMassProfilesList()
        
    return result

def combineToKeywordArgs(darkMatterFilamentList):
    '''
        Function to combine a all point masses in list of DarkMatterFilament objects
        to a keyword dictionary that can be used as keyword arguments for masses when
        creating a 'galaxy'.
        --------------------------------------------------------------------------------
        input: darkMatterFilamentList -> list of darkMatterFilament objects
        output: dictionary of all point masses in given darkMatterFilament objects
    '''
    combined_list = combine(darkMatterFilamentList)
    
    result = {}
    for i, mp in enumerate(combined_list):
        result["mass" + str(i)] = mp
    
    return result
```

# 4. Simulating Strong Gravitational Lensing by Dark Matter Filaments

#### 4.1 Image plane grid 

```python
# grid of size (100 x 100) and substructure size (2 x 2)
image_plane_grid = al.grid.uniform(
    shape_2d=(100, 100), pixel_scales=0.05, sub_size=2
)
```

#### 4.2 Lens composed of dark matter filaments

```python
# Dark matter filament with centre (0,0), oriented at 30 degrees and no jitter
dmf = DarkMatterFilament(centre = (0,0), angle = 30, jitter = False)

# Dark matter filament with centre (0.2,0), oriented at -80 degrees and with jitter
dmf2 = DarkMatterFilament(centre = (0.2,0), angle = -80, jitter = True)

# get dictionary of all point masses in the two dark matter filaments just defined
all_point_masses = combineToKeywordArgs([dmf, dmf2])

# create lens composed of the two dark matter filaments defined above
lens = al.Galaxy(redshift=0.5, **all_point_masses)

print("Lens:\n", lens)
```

#### 4.3 Source galaxy

```python
# Source galaxy with a standard sperical sersic light profile
sersic_light_profile = al.lp.SphericalSersic(
    centre=(0.0, 0.0), intensity=1.0, effective_radius=1.0, sersic_index=1.0
)
# define source galaxy
source_galaxy = al.Galaxy(redshift=1.0, light=sersic_light_profile)
print("Source Galaxy:\n", source_galaxy)
```

#### 4.4 Tracing

```python
# define tracer object with lens and source galaxy
tracer = al.Tracer.from_galaxies(galaxies=[lens, source_galaxy])
```

```python
traced_grids = tracer.traced_grids_of_planes_from_grid(grid = image_plane_grid)

# plot image plane grid
aplt.plane.plane_grid(
    plane=tracer.image_plane,
    axis_limits=[-2.5, 2.5, -2.5, 2.5],
    grid=traced_grids[0], 
    plotter=aplt.Plotter(labels=aplt.Labels(title="Image-plane Grid")),
)

# plot source plane grid traced back from image plane
aplt.plane.plane_grid(
    plane=tracer.source_plane,
    axis_limits=[-5.0, 5.0, -5.0, 5.0],
    grid=traced_grids[1], 
    plotter=aplt.Plotter(labels=aplt.Labels(title="Source-plane Grid")),
)

# source plane grid zoomed in
aplt.plane.plane_grid(
    plane=tracer.source_plane,
    grid=traced_grids[1],
    axis_limits=[-1, 1, -1, 1],
    plotter=aplt.Plotter(labels=aplt.Labels(title="Source-plane Grid")),
)
```

```python
# profile image of image plane (image to be simulated)
aplt.tracer.profile_image(tracer=tracer, grid=image_plane_grid)
```

#### 4.4 Simulating

```python
# Point Spread Function to simulate a telescopes optice
psf = al.kernel.from_gaussian(shape_2d = (11, 11), sigma = 0.1, pixel_scales = 0.1)

# Pad the profile image to prevent edge detects
normal_profile_image = tracer.profile_image_from_grid(grid = image_plane_grid)
padded_profile_image = tracer.padded_profile_image_from_grid_and_psf_shape(
    grid = image_plane_grid, psf_shape_2d = psf.shape_2d
)

# Add telescope optics, background sky and possion noice to simulated image
simulator = al.simulator.imaging(
    shape_2d = image_plane_grid.shape_2d,
    pixel_scales = 0.1,
    sub_size = image_plane_grid.sub_size,
    exposure_time = 300.0,
    psf = psf,
    background_level = 0.1,
    add_noise = True,
)

# Plot the simulated image
imaging = simulator.from_tracer(tracer = tracer)
aplt.imaging.image(imaging = imaging)
```

```python
# some extra plots
aplt.imaging.subplot_imaging(imaging = imaging)
```
