```python
import autolens as al
import autolens.plot as aplt

from astropy import cosmology
```

```python
grid = al.grid.uniform(
    shape_2d=(200, 200), pixel_scales=0.025, sub_size=4
)
```

```python
print(grid.sub_shape_2d)
print(grid.sub_shape_1d)
```

```python
mass_profile1 = al.mp.EllipticalIsothermal(centre=(-0.8, -0.8), einstein_radius = 1.6, axis_ratio=0.8, phi=90.0)
mass_profile2 = al.mp.EllipticalIsothermal(centre=(-0.4, -0.4), einstein_radius = 1.6, axis_ratio=0.8, phi=90.0)
mass_profile3 = al.mp.EllipticalIsothermal(centre=(0.0, 0.0),einstein_radius = 1.6, axis_ratio=0.8, phi=90.0)
mass_profile4 = al.mp.EllipticalIsothermal(centre=(0.4, 0.4), einstein_radius = 1.6, axis_ratio=0.8, phi=90.0)
mass_profile5 = al.mp.EllipticalIsothermal(centre=(0.8, 0.8), einstein_radius = 1.6, axis_ratio=0.8, phi=90.0)
light_prof1 = al.lp.EllipticalSersic(centre = (0.0,0.0), intensity=1.0, effective_radius= 1.0, sersic_index=2.0)
shear_prof =al.mp.ExternalShear(magnitude=0.05, phi=45.0)
```

```python
lens_galaxy = al.Galaxy(
    redshift=0.5,
    light=al.lp.SphericalSersic(
        centre=(0.0, 0.0), intensity=2.0, effective_radius=0.5, sersic_index=2.5
    ),
    mass_1 = mass_profile1,
    mass_2 = mass_profile2,
    mass_3 = mass_profile3,
    mass_4 = mass_profile4,
    mass_5 = mass_profile5
    ,
    shear = shear_prof)
    

print(lens_galaxy)
```

```python
lens_satellite = al.Galaxy(
    redshift=0.5,
    light=al.lp.SphericalDevVaucouleurs(
        centre=(1.0, 0.0), intensity=2.0, effective_radius=0.2
    ),
    mass=al.mp.SphericalIsothermal(centre=(1.0, 0.0), einstein_radius=0.4),
)

print(lens_satellite)
```

```python
aplt.galaxy.profile_image(
    galaxy=lens_galaxy,
    grid=grid,
    plotter=aplt.Plotter(labels=aplt.Labels(title="Lens Galaxy")),
)

aplt.galaxy.profile_image(
    galaxy=lens_satellite,
    grid=grid,
    plotter=aplt.Plotter(labels=aplt.Labels(title="Lens Satellite")),
)
```

```python
aplt.galaxy.deflections_y(
    galaxy=lens_galaxy,
    grid=grid,
    plotter=aplt.Plotter(labels=aplt.Labels(title="Lens Galaxy Deflections (y)")),
)
aplt.galaxy.deflections_y(
    galaxy=lens_satellite,
    grid=grid,
    plotter=aplt.Plotter(labels=aplt.Labels(title="Lens Satellite Deflections (y)")),
)
aplt.galaxy.deflections_x(
    galaxy=lens_galaxy,
    grid=grid,
    plotter=aplt.Plotter(labels=aplt.Labels(title="Lens Galalxy Deflections (x)")),
)
aplt.galaxy.deflections_x(
    galaxy=lens_satellite,
    grid=grid,
    plotter=aplt.Plotter(labels=aplt.Labels(title="Lens Satellite Deflections (x)")),
)
```

```python
tracer = al.Tracer.from_galaxies(
    galaxies=[lens_galaxy, lens_satellite, source_galaxy_0],
    cosmology=cosmology.Planck15,
)
```

```python
aplt.tracer.profile_image(tracer=tracer, grid=grid)
```

```python
traced_grids = tracer.traced_grids_of_planes_from_grid(grid=grid)

aplt.plane.plane_grid(
    plane=tracer.source_plane, 
    grid=traced_grids[1], 
    plotter=aplt.Plotter(labels=aplt.Labels(title="Source-plane Grid")),
)
```

```python
aplt.tracer.subplot_tracer(tracer=tracer, grid=grid)
```

```python
psf = al.kernel.from_gaussian(shape_2d=(11, 11), sigma=0.1, pixel_scales=0.1)
```

```python
normal_image = tracer.profile_image_from_grid(grid=grid)
padded_image = tracer.padded_profile_image_from_grid_and_psf_shape(
    grid=grid, psf_shape_2d=psf.shape_2d
)
print(normal_image.shape)
print(padded_image.shape)
```

```python
simulator = al.simulator.imaging(
    shape_2d=grid.shape_2d,
    pixel_scales=0.1,
    sub_size=grid.sub_size,
    exposure_time=300.0,
    psf=psf,
    background_level=0.1,
    add_noise=True,
)

imaging = simulator.from_tracer(tracer=tracer)

```

```python
aplt.imaging.image(imaging=imaging)
```
